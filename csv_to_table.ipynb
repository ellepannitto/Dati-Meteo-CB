```python
class Dato:
    def __init__(self,
                 stazione="NA",
                 giorno="NA",
                 mese="NA",
                 anno="NA",
                 temp_max="NA", 
                 temp_min="NA", 
                 temp_med="NA", 
                 hum_max="NA", 
                 hum_min="NA", 
                 hum_med="NA", 
                 rain_mm="NA", 
                 wind_int="NA", 
                 wind_dir="NA"):
        
        self.stazione=stazione
        self.date = (giorno, mese, anno)
        self.temp_max = temp_max
        self.temp_min = temp_min
        self.temp_med = temp_med
        self.hum_max = hum_max
        self.hum_min = hum_min
        self.hum_med = hum_med
        self.rain_mm = rain_mm
        self.wind_int = wind_int
        self.wind_dir = wind_dir
        #print(self.date)
        
    def __str__(self):
        return str(self.date)
```

```python
LISTA_DATI = []

mesi = ["", "gennaio", "febbraio", "marzo", "aprile", "maggio", "giugno", "luglio", 
        "agosto", "settembre", "ottobre", "novembre", "dicembre"]
```

```python
# READER 2017, 2018, 2019, 2020a

for anno, filename in zip([2017, 2018, 2019, 2020, 2021], 
                          ["dati_meteo - 2017.tsv", 
                           "dati_meteo - 2018.tsv", 
                           "dati_meteo - 2019.tsv",
                           "dati_meteo - 2020a.tsv",
                           "dati_meteo - 2021a.tsv"]):
    with open(filename) as fin:
        _ = fin.readline()

        dates = fin.readline().strip().split("\t")    
        dates = [x.split(" ")[-1] for x in dates if len(x)>0]

        stazioni = fin.readline().strip().split("\t")
        stazioni = [x.split(":")[-1] for x in stazioni if len(x)>0]
        stazioni = [x[:-1] if x.endswith(")") else x for x in stazioni]

        assert(len(dates) == len(stazioni))

        _ = fin.readline()
        _ = fin.readline()

        for newline in fin:
            newline = newline.strip().split("\t")

            #print(newline)        

            blocchi = []
            i=0
            while i<len(newline):
                blocchi.append(newline[i:i+8])
                i += 9
            
            #assert(len(dates) == len(blocchi))

            for i, blocco in enumerate(blocchi):
                #print(blocco)
                if not blocco[0] in ["", "Media"]:
                    giorno, temp_max, temp_min, temp_med, hum_med, rain_mm, wind_int, wind_dir = blocco
                    mese = mesi.index(dates[i].lower())
                    stazione = stazioni[i]

                    giorno = int(giorno)
                    temp_max = float(temp_max.replace(',','.').replace("‐", "-"))
                    temp_min = float(temp_min.replace(',','.').replace("‐", "-"))
                    temp_med = float(temp_med.replace(',','.').replace("‐", "-"))
                    hum_med = float(hum_med.replace(',','.').replace("‐", "-"))
                    rain_mm = float(rain_mm.replace(',','.').replace("‐", "-"))
                    wind_int = float(wind_int.replace(',','.').replace("‐", "-"))

                    nuovo_dato = Dato(stazione=stazione,
                                      giorno=giorno,
                                      mese=mese,
                                      anno=anno,
                                      temp_max=temp_max,
                                      temp_min=temp_min,
                                      temp_med=temp_med,
                                      hum_med=hum_med,
                                      rain_mm=rain_mm,
                                      wind_int=wind_int,
                                      wind_dir=wind_dir)

                    print(nuovo_dato)
                    LISTA_DATI.append(nuovo_dato)
```

```python
# READER 2020b

for anno, filename in zip([2020], 
                          ["dati_meteo - 2020b.tsv" ]):
    with open(filename) as fin:
        _ = fin.readline()

        stazioni = fin.readline().strip().split("\t")
        stazioni = [x for x in stazioni if len(x)>0]

        _ = fin.readline()


        for newline in fin:
            newline = newline.strip().split("\t")

            #print(newline)        

            blocchi = []
            i=0
            while i<len(newline):
                blocchi.append(newline[i:i+9])
                i += 10
            
            #assert(len(dates) == len(blocchi))

            for i, blocco in enumerate(blocchi):
                #print(blocco)
                if not blocco[0] in ["", "Media"]:
                    data, temp_min, temp_max, temp_med, hum_min, hum_max, hum_med, wind_int, rain_mm = blocco
                    #print(data)
                    
                    try:
                        giorno, mese, _ = data.split("/")
                    except:
                        giorno, mese = data.split("/")
                    
                    mese = int(mese)
            
                    stazione = stazioni[i]

                    giorno = int(giorno)
                    temp_max = float(temp_max.replace(',','.').replace("‐", "-"))
                    temp_min = float(temp_min.replace(',','.').replace("‐", "-"))
                    temp_med = float(temp_med.replace(',','.').replace("‐", "-"))
                    hum_min = float(hum_min.replace(',','.').replace("‐", "-"))
                    hum_max = float(hum_max.replace(',','.').replace("‐", "-"))
                    hum_med = float(hum_med.replace(',','.').replace("‐", "-"))
                    rain_mm = float(rain_mm.replace(',','.').replace("‐", "-"))
                    wind_int = float(wind_int.replace(',','.').replace("‐", "-"))

                    nuovo_dato = Dato(stazione=stazione,
                                      giorno=giorno,
                                      mese=mese,
                                      anno=anno,
                                      temp_max=temp_max,
                                      temp_min=temp_min,
                                      temp_med=temp_med,
                                      hum_med=hum_med,
                                      hum_min=hum_min,
                                      hum_max=hum_max,
                                      rain_mm=rain_mm,
                                      wind_int=wind_int)

                    print(nuovo_dato)
                    LISTA_DATI.append(nuovo_dato)
```

```python
# READER 2021b

mesi = ["", "gennaio", "febbraio", "marzo", "aprile", "maggio", "giugno", "luglio", 
        "agosto", "settembre", "ottobre", "novembre", "dicembre"]

for anno, filename in zip([2021], 
                          ["dati_meteo - 2021b.tsv" ]):
    with open(filename) as fin:
        _ = fin.readline()

        dates = fin.readline().strip().split("\t")    
        dates = [x.split(" ")[-1] for x in dates if len(x)>0]
        
        stazioni = fin.readline().strip().split("\t")
        stazioni = [x for x in stazioni if len(x)>0]

        _ = fin.readline()
        _ = fin.readline()


        for newline in fin:
            newline = newline.strip().split("\t")

            #print(newline)        

            blocchi = []
            i=0
            while i<len(newline):
                blocchi.append(newline[i:i+10])
                i += 11
            
            #assert(len(dates) == len(blocchi))

            for i, blocco in enumerate(blocchi):
                #print(blocco)
                if not blocco[0] in ["", "Media"]:
                    giorno, temp_max, temp_min, temp_med, hum_med, rain_mm, wind_int, wind_dir, hum_max, hum_min = blocco
                    mese = mesi.index(dates[i].strip("?").lower())
                    stazione = stazioni[i]

                    giorno = int(giorno)
                    temp_max = float(temp_max.replace(',','.').replace("‐", "-"))
                    temp_min = float(temp_min.replace(',','.').replace("‐", "-"))
                    temp_med = float(temp_med.replace(',','.').replace("‐", "-"))
                    hum_med = float(hum_med.replace(',','.').replace("‐", "-"))
                    rain_mm = float(rain_mm.replace(',','.').replace("‐", "-"))
                    wind_int = float(wind_int.replace(',','.').replace("‐", "-"))
                    hum_max = float(hum_max.replace(',','.').replace("‐", "-"))
                    hum_min = float(hum_min.replace(',','.').replace("‐", "-"))

                    nuovo_dato = Dato(stazione=stazione,
                                      giorno=giorno,
                                      mese=mese,
                                      anno=anno,
                                      temp_max=temp_max,
                                      temp_min=temp_min,
                                      temp_med=temp_med,
                                      hum_med=hum_med,
                                      rain_mm=rain_mm,
                                      wind_int=wind_int,
                                      wind_dir=wind_dir,
                                      hum_max=hum_max,
                                      hum_min=hum_min)

                    print(nuovo_dato)
                    LISTA_DATI.append(nuovo_dato)
```

```python
dati_sorted = sorted(LISTA_DATI, key=lambda x: (x.date[2], x.date[1], x.date[0]))
with open("serie_meteo.tsv", "w") as fout:
    print("Giorno\tMese\tAnno\tStazione\tTemp_max\tTemp_min\tTemp_med\tHum_max\tHum_min\tHum_med\tRain_mm\tWind_int\tWind_dir", file=fout)
    for dato in dati_sorted:        
        print("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}".format(dato.date[0], 
                                                                    dato.date[1], 
                                                                    dato.date[2],
                                                                    dato.stazione,
                                                                    dato.temp_max,
                                                                    dato.temp_min,
                                                                    dato.temp_med,
                                                                    dato.hum_min,
                                                                    dato.hum_max,
                                                                    dato.hum_med,
                                                                    dato.rain_mm,
                                                                    dato.wind_int,
                                                                    dato.wind_dir), file=fout)
```
